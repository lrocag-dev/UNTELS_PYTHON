## Pregunta 2
Se piden tres notas y luego pregunte si se ingresa una nota sustitutoria, si es así entonces esa nota reemplaza la mas baja de las tres primeras, y luego se muestra el promedio de las tres notas.


```python
[a,b,c]=[int(x) for x in input("Ingrese 3 notas: ").split()]
```

```python
rpta = input("Desea ingresar nota sustitutoria? (Si/No): ")
if rpta.lower()=='s':
    s=int(input("Ingrese nota sustitutoria: "))
    prom=(a+b+c-min(a,b,c)+s)/3
else:
    prom=(a+b+c)/3
print(f"Promedio: {prom:5.1f}")
```

## Pregunta 3
Se piden 6 números reales $(a,fa)$, $(b,fb)$, $(c,fc)$ con la condición de que $a<b<c$, estos pares definen segmentos de recta, y luego pida un número $x$. El algoritmo debe mostrar el valor de $y$ correspondiente.

```python
[a,fa]=[float(x) for x in input("Ingrese (a, fa): ").split()]
[b,fb]=[float(x) for x in input("Ingrese (b, fb): ").split()]
[c,fc]=[float(x) for x in input("Ingrese (c, fc): ").split()]
x = float(input("Ingrese x en [a,c]: "))
if (a<=x<=b):
    y = fa + (x-a)*(fb-fa)/(b-a)
if (b<=x<=c):
    y = fb + (x-b)*(fc-fb)/(c-b)
print(f"y={y:5.2f}")
```

## Pregunta 4
Se piden el área de un terreno y el precio por $m^2$ y luego se muestra el impuesto predial según el siguiente cuadro

|Tasa | Area ($m^2$)|
|---- | ----------- |
|2 \% | 0 - 90      |
|5 \% | 91 - 200    |
|8 \% | 201 - 1000  |
|11 \% | $\ge$ 1001 |


```python
area = int(input("Area m2: "))
preciom2 = float(input("Precio m2: "))

if 0<=area<=90:
    tasa=0.02
elif 91<=area<=200:
    tasa=0.05
elif 201<=area<=1000:
    tasa=0.08
else:
    tasa=0.11

impuesto = tasa*area*preciom2
print(f"Impuesto predial {impuesto:5.3f}")
```

Solución con estructura repetitiva


```python
area = int(input("Area m2: "))
preciom2 = float(input("Precio m2: "))
tasas = [0.02,0.05,0.08,0.11]
lim = [90,200,1000]
for icaso in range(len(lim)):
    if area <= lim[icaso]:
        tasa = tasas[icaso]
        break
if area>lim[-1]:
    tasa = tasas[-1]
impuesto = tasa*area*preciom2
print(f"Impuesto predial {impuesto:5.3f}")
```

## Pregunta 5
Se piden  8 números reales: $(a,pa)$, $(b,pb)$, $(c,pc)$ vértices de un triángulo y $(x,y)$ un punto del plano, y luego determine cual vértice es el mas cercano.

```python
[a,pa]=[float(x) for x in input("Ingrese (a, pa): ").split()]
[b,pb]=[float(x) for x in input("Ingrese (b, pb): ").split()]
[c,pc]=[float(x) for x in input("Ingrese (c, pc): ").split()]
[x,y]=[float(x) for x in input("Ingrese (x, y): ").split()]
from math import sqrt
da=sqrt( (x-a)**2 + (y-pa)**2)
db=sqrt( (x-b)**2 + (y-pb)**2)
dc=sqrt( (x-c)**2 + (y-pc)**2)
if (da<=db) and (da<=dc):
    print(f"Punto mas cercano ({a},{pa})")
elif (dc<=db):
    print(f"Punto mas cercano ({c},{pc})")
else :
    print(f"Punto mas cercano ({b},{pb})")
```

## Problema 6
Calcula el volumen de agua que contiene un reciente que tiene la forma de una semiesfera de radio $R$ en la parte inferior y un cono circular recto de altura $H$ en la parte superior si se conoce la altura $h$ del agua.

```python
from math import pi
R=float(input("Radio de la semiesfera: "))
H=float(input("Altura del cono circular: "))
h=float(input("Altura del agua: "))
if 0<=h<=R: #estamos en la semiesfera
    volumen = pi/3 * (3*R-h) * h**2
elif R<=h<=R+H: #estamos en el cono
    volumen = 2/3*pi*R**3 + 1/3*pi*R**2*H-1/3*pi*(R*(H-h+R)/H)**2*(H-h+R);
print(f"Volumen = {volumen:5.2f}")
```
