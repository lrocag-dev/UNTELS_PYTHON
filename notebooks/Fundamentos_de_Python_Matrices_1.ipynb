````python
trace_A = np.trace(A)
det_A = np.linalg.det(A)
I = np.eye(2)
A @ A - trace_A * A + det_A * I
````

## Ejercicio

Implemente las transformaciones rigidas (https://www.matem.unam.mx/~max/GEA16/N11.pdf) como funciones que acepten un array (x,y,1) y devuelvan un array (x',y',1)

1.- La traslacion

$$
T_h(x,y)= \begin{bmatrix}
1 &0 &h_1  \\
0 &1 &h_2 \\
0 &0& 1 \\
\end{bmatrix}
\begin{bmatrix}
x\\
y \\
1 \\
\end{bmatrix}
$$

````python
def T(x,h):
    h1 = h[0]
    h2 = h[1]
    matT=np.array([ [1,0,h1], [0,1,h2] , [0,0,1] ])
    z  = matT @ x
    return z
````

````python
x = np.array([2,1,1])
h1 = np.array([1,3,1])
h2 = np.array([2,2,1])
print(T(x,h1))
````

````python
def mT(h):
    h1 = h[0];  h2 = h[1]
    matT=np.array([ [1,0,h1], [0,1,h2] , [0,0,1] ])
    return matT
````

````python
mT(h1) @ x
````

````python
mT(h2) @ mT(h1) @ x
````

````python
mT(h2+h1) @ x
````

````python
mT(-h1) @ mT(h1) @ x
````

2.- La rotación

$$
R_\theta(x,y)= \begin{bmatrix}
\cos \theta & -\sin \theta &0  \\
\sin \theta & \cos \theta &0 \\
0 &0& 1 \\
\end{bmatrix}
\begin{bmatrix}
x\\
y \\
1 \\
\end{bmatrix}
$$

````python
def rotacion(x,theta):
    ct=np.cos(theta)
    st=np.sin(theta)
    matR=np.array([ [ct,-st,0], [st,ct,0] , [0,0,1] ])
    z  = matR @ x
    return z
````

````python
x = np.array([2,1,1])
theta = np.pi/2
print(rotacion(x,theta))
````

````python
def mR(theta):
    ct=np.cos(theta)
    st=np.sin(theta)
    matR=np.array([ [ct,-st,0], [st,ct,0] , [0,0,1] ])
    return matR
````

````python
mR(np.pi/2) @ x
````

````python
mR(np.pi/3)  @ mR(np.pi/2) @ x
````

````python
mR(np.pi/3 + np.pi/2) @ x
````
