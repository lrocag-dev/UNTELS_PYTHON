# Gráficas en matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Conceptos
* pyplot es un módulo, contiene funciones de dibujo.
* Figure es una clase, se refiere a la ventana donde dibujamos las graficas, comprende: titulo, subtitulo, legenda, barra de color y Axes.
* Axes es una clase, es el area donde se crea la grafica, un Figure pueden contener multiples Axes.
* Cada metodo de dibujo de Axes tiene su función equivalente en pyplot.

### Ejemplo

```python
fig = plt.figure()
ax = fig.add_subplot(111)
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot([1, 2, 3, 4], [10, 20, 25, 30], color='lightblue', linewidth=3)
ax.scatter([0.3, 3.8, 1.2, 2.5], [11, 25, 9, 26], color='darkgreen', marker='^')
ax.set_xlim(0.5, 4.5)
plt.show()
```

Cuand usamos pyplot, el dibujo afecta al Axes activo

```python
plt.plot([1, 2, 3, 4], [10, 20, 25, 30], color='lightblue', linewidth=3)
plt.scatter([0.3, 3.8, 1.2, 2.5], [11, 25, 9, 26], color='darkgreen', marker='^')
plt.xlim(0.5, 4.5)
plt.show()
```

## Graficas en dos dimensiones

```python
x=np.array([1, 2, 3, 4, 5])
y=np.array([10, 5, 8, 15, 20])
```

```python
font = {'family':'serif','color':'darkred','size':16 }
plt.plot(x,y,'*-b',label='curva 1')
plt.plot(x,20-y,'+--g',label='curva 2')
plt.plot(x,10-2*y,'^:r',label='curva 3')
plt.plot(x,5+y,'k-.D',label='curva 4')
plt.legend(loc='lower center')
plt.xlabel('x',fontdict=font)
plt.ylabel('y',fontdict=font)
plt.text(1.5,-20,r'$y=f(x)$',fontdict=font)
plt.grid('on')
plt.title('Gráfica 1',fontdict=font,loc='left')
```

## Varias graficas en una figura

```python
x=np.linspace(1,5,10)
plt.subplot(2,2,1)
y1=x*(np.sin(x)+1)
plt.plot(x,y1,'r')
plt.title('y=x(sen(x)+1)')
plt.subplot(2,2,2)
y2=np.sin(x/2)/x
plt.plot(x,y2,'g')
plt.title('y=sen(x/2) / x')
plt.subplot(2,2,3)
plt.plot(x,1/x,'c')
plt.title('y=1/x')
plt.subplot(2,2,4)
plt.plot(x,np.cos(1/x),'magenta')
plt.title('y=cos(1/x)')
plt.suptitle('Grafica 2')
plt.subplots_adjust(top=0.85, bottom=0.08, left=0.1,right=0.95,
                   hspace=0.35,wspace=0.25)
# hspace ->  altura de la separacion
# wspace ->  ancho de la separacion
```

```python
fig,ax = plt.subplots(2,2,figsize=(6,5),
                      gridspec_kw={'top':0.85, 'bottom':0.08, 'left':0.1,'right':0.95,
                   'hspace':0.35,'wspace':0.2})
ax[0,0].plot(x,y1,'r')
ax[0,0].set_title('y=x(sen(x)+1)')
ax[0,1].plot(x,y2,'g')
ax[0,1].set_title('y=sen(x/2) / x')
fig.suptitle('Grafica 2')
```

## Grafica de contornos

```python
dx, dy = 0.05, 0.05
y, x = np.mgrid[slice(1, 5 + dy, dy),
                slice(1, 5 + dx, dx)]

z = np.sin(x)**10 + np.cos(10 + y*x) * np.cos(x)
cmap = plt.get_cmap('PiYG')

fig, (ax0, ax1) = plt.subplots(nrows=2,figsize=(15,15))

im = ax0.pcolormesh(x, y, z, cmap=cmap)
fig.colorbar(im, ax=ax0)
ax0.set_title('pcolormesh')

cf = ax1.contourf(x,y,z,cmap=cmap)
fig.colorbar(cf, ax=ax1)
ax1.set_title('contourf')

# ajusta el espacio entre graficas
fig.tight_layout()
```

```python
dx, dy = 0.3, 0.3
y, x = np.mgrid[slice(1, 5 + dy, dy),
                slice(1, 5 + dx, dx)]

z = np.sin(x)**10 + np.cos(10 + y*x) * np.cos(x)
cmap = plt.get_cmap('PiYG')

fig, (ax0, ax1) = plt.subplots(nrows=2)

im = ax0.pcolormesh(x, y, z, cmap=cmap)
fig.colorbar(im, ax=ax0)
ax0.set_title('pcolormesh')

cf = ax1.contourf(x,y,z,cmap=cmap)
fig.colorbar(cf, ax=ax1)
ax1.set_title('contourf')

# ajusta el espacio entre graficas
fig.tight_layout()
```

## Gráficas en tres dimensiones

```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm

fig = plt.figure()
ax = fig.gca(projection='3d')

# Make data.
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

# Plot the surface.
surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)

# Customize the z axis.
ax.set_zlim(-1.01, 1.01)

# Add a color bar which maps values to colors.
fig.colorbar(surf, shrink=0.5, aspect=5)

plt.show()
```

## Curvas paramétricas

$\mathcal{C}: x=x(t), y=y(t)$

### Ejemplo

Graficar $r^2=sen(2t), t\in [0,\pi/2]$

```python
import numpy as np
import matplotlib.pyplot as plt

t=np.linspace(0,np.pi/2,100)
x = np.sqrt(np.sin(2*t))*np.cos(t)
y = np.sqrt(np.sin(2*t))*np.sin(t)
plt.plot(x,y)
plt.plot(-x,-y)
plt.gca().set_aspect('equal')
```

### Ejemplo

Graficar $r=z^2+1$ (Coordenadas cilíndricas)

```python
from mpl_toolkits.mplot3d import Axes3D

plt.rcParams['legend.fontsize'] = 10

fig = plt.figure()
ax = fig.gca(projection='3d')

theta = np.linspace(-4 * np.pi, 4 * np.pi, 100)
z = np.linspace(-2, 2, 100)
r = z**2 + 1
x = r * np.sin(theta)
y = r * np.cos(theta)

ax.plot(x, y, z, label='curva 3d')
ax.legend()
plt.savefig("/usr/local/share/jupyter/nbextensions/curva1.png", transparent=True, dpi=150)
plt.savefig("/usr/local/share/jupyter/nbextensions/curva2.png",  transparent=True,dpi=50)
plt.show()
```

## Extension del mapa
Utilizamos la función set_extent

```python
fig = plt.figure(figsize=(10, 5))
ax = fig.add_subplot(1, 1, 1, projection=ccrs.PlateCarree())
ax.stock_img()
ax.coastlines()
ax.set_extent([-85, -60, 0,-20], crs=ccrs.PlateCarree())
plt.show()
```

## Personalizar un mapa

```python
import cartopy.feature as cfeature
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(1, 1, 1, projection=ccrs.PlateCarree())
ax.set_extent([-85, -60, 0,-20], crs=ccrs.PlateCarree())

ax.add_feature(cfeature.LAND)
ax.add_feature(cfeature.OCEAN)
ax.add_feature(cfeature.COASTLINE)
ax.add_feature(cfeature.BORDERS, linestyle=':')
ax.add_feature(cfeature.LAKES, alpha=0.5)
ax.add_feature(cfeature.RIVERS)

plt.show()
```
