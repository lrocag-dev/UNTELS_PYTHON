# Problema 1

Defina una funcion que evalue un numero real $x$ entre 0 y 1, y devuelva
$$1+x+x^2+\cdots+x^n$$



```python
def problema1(x,n):
  if 0<x<1:
    s = 1
    t = 1
    for i in range(n):
      t = t*x
      s += t
  else:
    s = None
  return s
```

```python
problema1(0.5,10)
```

# Problema 3
Un estacionamiento cobre 2.5 soles por hora o fraccion. Escriba un función que devuelva el monto a pagar si se ingresa el timepo en minutos.

```python
def problema3(t):
  n = t // 60
  r = t % 60
  if r>0:
    n = n+1
  monto = n*2.5
  return monto
```

```python
problema3(120)
```

```python
problema3(70)
```

# Problema 4

Escriba una función que pida el ingreso de notas, el examen parcial, final y promedio de prácticas, laboratorio, tarea académica. Tambien pregunta si desea ingrasar nota sustitutoria que remeplaza a la nota mas baja entre el examen parcial y el final. La funcion devuelve el promedio del curso.

```python
def problema4():
  notas = {"EF":None, "ES":None, "PP":None, "EP":None,
           "LAB":None, "TA":None}
  pesos = {"EF":0.25, "ES":0.25, "PP":0.2, "EP":0.25,
           "LAB":0.2, "TA":0.1}
  for x in notas:
    if x!="ES":
      notas[x] = float(input(f"Ingrese {x} : "))
  opc = input(f"¿Desea ingresar nota sustitutoria? (S)i (N)o: ")
  if opc.upper() == "S":
    notas["ES"]  = float(input(f"Ingrese ES : "))
  s = 0
  for x in notas:
    if notas[x] is not None:
      s = s + notas[x]*pesos[x]
  if notas["ES"] is not None:
      s = s - 0.25*min(notas["EF"],notas["EP"])
  return s
```

# Problema 5
Una funcion pide el ingreso de caracteres y devuelve el números de veces que se ingreso una vocal.

```python
def problema5():
  c=''
  vocales = {'a':0,'e':0,'i':0,'o':0,'u':0}
  while c!='x':
    c = input("Ingrese un caracter: ")
    c = c.lower()
    if c in vocales:
      vocales[c] += 1
  for c in vocales:
    print(f"La vocal {c} se ingreso {vocales[c]} veces")

```

```python
problema5()
```
