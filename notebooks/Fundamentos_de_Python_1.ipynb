## Version de Python

```python
import sys
```

```python
print(sys.version_info)
```

Se utiliza ampliamente en el código para verificar la compatibilidad de versiones sys de Python. Por ejemplo:


```python
print(f"Su version de Python es  {sys.version}")
if sys.version_info >= (3, 10):
  print("Código para Python 3.10 o superior.")
else:
  print(f"Su version de Python está obsoleta.")
  pass
```
