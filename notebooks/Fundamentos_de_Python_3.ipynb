## Diccionarios por comprensión

```python
cuadrados = {x:x*x for x in range(1,11)}
```

```python
print(cuadrados)
```

```python
cuadrados_impares = {x:x*x for x in range(1,11) if x%2 !=0}
```

```python
print(cuadrados_impares)
```

```python
lista1 = [3,4,5,11,20]
```

```python
cuadrados_impares_2 = {x:x*x for x in range(1,11) if x%2 !=0 and x not in lista1}
```

```python
print(cuadrados_impares_2)
```

```python
cuadrados_impares_3 = {x:(x*x if x%2==0 else x*x*x) for x in range(1,11) }
```

```python
print(cuadrados_impares_3)
```

## Diccionarios anidados

```python
d2 = { x : {y :x*y for y in range(1,6)} for x in range(3,6) }
```

```python
print(d2)
```

```python
3 in cuadrados_impares
```

```python
for x in cuadrados_impares:
    print(cuadrados_impares[x])
```

### Ejemplo
Calcular la suma de suadrados de una lista

```python
x = [1]*100
```

Un microsegundo equivale a 10⁻⁶ segundos, mientras que un nanosegundo equivale a 10⁻⁹ segundos.

```python
%%timeit
suma=0
for val in x:
    suma+=x[val]
```

```python
%%timeit
sum(x)
```
