## 9. Create a color plot of  $\sin(x)\sin(y)$  on the interval  $[-\pi,\pi]\times [-\pi,\pi]$ .

````python
X, Y = np.meshgrid(np.linspace(-np.pi, np.pi, 100),
                   np.linspace(-np.pi, np.pi, 100))
````

````python
X.shape
````

````python
Z = np.sin(X) * np.sin(Y)
````

````python
fig = plt.figure()
ax = plt.axes()
ax.set_xlim(-np.pi, np.pi)
ax.set_ylim(-np.pi, np.pi)
c = ax.pcolormesh(X, Y, Z, cmap=plt.get_cmap('PiYG'))
fig.colorbar(c)
````

````python
fig = plt.figure()
ax = plt.axes()
ax.set_xlim(-np.pi, np.pi)
ax.set_ylim(-np.pi, np.pi)
c = ax.contourf(X, Y, Z, cmap=plt.get_cmap('PiYG'))
fig.colorbar(c)
````

````python
from mpl_toolkits import mplot3d
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z)
ax.view_init(elev=45., azim=30)
#ax.plot_surface(X, Y, Z,cmap='viridis', edgecolor='none')
ax.set_title('Surface plot')
plt.show()
````

````python
def collatz(n):
    nOne = n == 1
    nEve = np.mod(n,2) == 0
    nOdd = nOne ^ (np.mod(n,2) == 1)

    return (nOne * 1 + nEve * (n / 2) + nOdd * (3 * n + 1))
````

````python
print(collatz(collatz(collatz(8))))
````

````python
print( collatz(15) )
````
