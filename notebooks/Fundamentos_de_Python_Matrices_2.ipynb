## Funciones para ordenamiento

```python
print(A)
np.sort(A,axis=0)
```

```python
np.sort(A,axis=1)
```

```python
np.sort(A,axis=1,kind='quicksort') # O(n^2)
```

```python
np.sort(A,axis=1,kind='heapsort') # O(n log(n))
```

```python
# columnas con nombre
dtype = [ ('nombre','S10'), ('peso','f4'), ('edad','d')]
values = [ ('Pepe', 68.2,28), ('Paco', 81.3,32), ('Paul', 71.3,26)]
```

```python
tabla = np.array(values, dtype = dtype)
print(tabla)
```

```python
np.sort(tabla, order='edad')
```

```python
np.sort(tabla, order=['edad','peso'])
```

```python
A,np.sort(A,axis=1),np.argsort(A)
```
