# Estructuras de decisión

Ejemplo: Determinar el menor de 3 numeros $x$,$y$,$z$

```python
x = float ( input ('x='))
y = float ( input ('y='))
z = float ( input ('z='))
if x < y and x < z:
  print ('x es el menor ')
elif y < z:
  print ('y es el menor ')
else :
  print ('z es el menor ')
```

Ejemplo: Determinar el mayor impar de 3 numeros $x$, $y$, $z$



```python
x = int ( input ('x='))
y = int ( input ('y='))
z = int ( input ('z='))
mayor = None
if x %2!=0 :
  if mayor is not None:
    if x > mayor :
      mayor = x
  else:
    mayor = x
if y %2!=0 :
  if mayor is not None:
    if y > mayor :
      mayor = y
  else:
    mayor = y
if z %2!=0 :
  if mayor is not None:
    if z > mayor :
      mayor = z
  else:
    mayor = z
if mayor is not None:
  print(f'El mayor impar es {mayor}')
else :
  print('Ninguno es impar ')
```

# Uso de cadenas

```python
lon = input ("ingrese longitud E-O :")
lon = lon.upper()# convierte en mayúsculas
if "O" in lon :
  sufijo ="O"; signo =-1
elif "E" in lon:
  sufijo ="E"; signo = 1
else :
  sufijo = None ; signo = 1

if sufijo is not None :
  lon = lon [0: lon.index ( sufijo )]
  lon = signo * float (lon )
print (lon)
```
